# Statistics of the data
Get the statistic data of the dataset with ABP Mean.
Each subject contains different types of signals. Find out which kinds of signals are contained in most of the subjects. Then use those signals for prediction.

```python
import shutil
import wfdb
import os
import numpy as np
from IPython.display import display

from collections import Counter
import logging
```

replace different part with "num"

```python
num = '35'
rootpath = '/research/dept7/xbzheng/mimic2wdb/'
rootpath = rootpath + num + '/'
rootrds = rootpath + 'RECORDS'
print(rootrds)
f = open(rootrds, "r")

allsigname = [] # to save all the sig_name in one list

# record all the subject path
fname_ABPrds = '/research/dept7/xbzheng/' + num + '_ABPrds.txt'
f_ABPrds = open(fname_ABPrds, 'a+')

while True:
# for i in range(1,10):        # just for testing
    n = f.readline()
    if not n:
        break
    folderpath = rootpath + n.strip('\n')
#     print(folderpath)
    subrds = folderpath + "RECORDS"
    f1 = open(subrds,"r")
    while True:
        n2 = f1.readline().strip('\n')
        if not n2:
            break
        if('n' in n2):
#             print(n2)
            subj = folderpath + n2
            print(subj)
            try:
                signals, fields = wfdb.rdsamp(subj)
            
                # get the signal if there is ABP mean in it. Delete 'if...' when you want to get all the signal
                if ('ABPMean' in fields['sig_name'] or 'ABP MEAN' in fields['sig_name'] or 'ABP Mean' in fields['sig_name'] ):
                    allsigname = allsigname + fields['sig_name']
                    print(fields['sig_name'])
                    f_ABPrds.write(subj + '\n')
            except Exception as e:
                logging.exception(e)

# print(allsigname)
f_ABPrds.close()
```

```python
result = Counter(allsigname)
print(result)
```

## Out put to csv

```python
import csv
filename = '/research/dept7/xbzheng/sig' + num + '_ABP.csv'
print(filename)
with open(filename,'w') as csvfile:
    fieldnames=['signal','count']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key in result.keys():
        csvfile.write("%s,%s\n"%(key,result[key]))
```
